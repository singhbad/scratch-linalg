```python
# Create 2 matrices using numpy array

A = np.array([[2,4,5],[8,3,6],[8,4,6]]) #Matrix A
B = np.array([[3,1,2],[4,9,0],[8,4,5]]) #Matrix B



```

```python
A.shape
```

```python
A
```

```python
#Sum Operation

print(A+B)
```

```python
#dIFFER ENCE Operation

print(A - B)
```

```python
#Finding product of the Matrices

#1st Way
print(A@B)

#2nd Way
print(np.matmul(A,B))
```

```python
#Finding transpose of a Matrix
#1St method
print(A.T)
```

```python
#2nd Method
print(np.transpose(A))

#Or

print(A.transpose())
```

```python
#Fidning determinant using built in method

print(np.linalg.det(A))
print(np.linalg.det(B))
```

```python
#Fidning Invesrse of a Matrix

print(np.linalg.inv(A))
print(np.linalg.inv(B))
```

```python
#Fidning Trace

print(np.linalg.trace(A))
print(np.linalg.trace(B))
```

```python
#Rank of a matrix

print(np.linalg.matrix_rank(A))
print(np.linalg.matrix_rank(B))
```

```python
#Fidning EigenValue and EigenVector

print(np.linalg.eig(A))
print(np.linalg.eig(B))
```
